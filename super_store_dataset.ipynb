```python
import matplotlib.pyplot as plt

# Assuming `data` is your DataFrame containing Superstore data
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Order Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
region_sales_profit = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales')

region_sales_profit.plot(kind='bar', figsize=(10, 6), color=['orange', 'green'])
plt.title('Total Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.legend(['Sales', 'Profit'])
plt.show()
```

```python
sns.boxplot(x='Ship Mode', y='Sales', data = df)
plt.title('Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
plt.show()

```

```python
sns.scatterplot(x='Sales', y='Profit', hue='Segment', data=df, palette='viridis')
plt.title('Customer Segmentation: Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Quantity'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Quantity Ordered')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
top_states_sales = df.groupby('State')['Sales'].sum().nlargest(10)

top_states_sales.plot(kind='barh', figsize=(10, 6), color='coral')
plt.title('Top 10 States by Sales')
plt.xlabel('Sales')
plt.ylabel('State')
plt.show()
```

```python
top_products = df.groupby('Product Name')[['Sales', 'Profit']].sum().nlargest(10, 'Sales')

top_products.plot(kind='bar', figsize=(12, 6))
plt.title('Top Products by Sales and Profit')
plt.xlabel('Product Name')
plt.ylabel('Amount')
plt.legend(['Sales', 'Profit'])
plt.show()
```
